```python
my_data = pd.DataFrame({"EnglishText":["my name is Amirul", "Good boy","Data Science Good Future","Proud to be 'AI Is Future'"]})
my_data
```

```python

from translate import Translator
import pandas as pd

# Assuming you have a DataFrame named 'my_data' with columns 'EnglishText' and 'TamilText'
translator = Translator(to_lang="ta")

# Use the apply function to translate the EnglishText column to Tamil
my_data["TamilText"] = my_data["EnglishText"].apply(lambda x: translator.translate(x))

# Display the DataFrame with the translated text
print(my_data)

```
